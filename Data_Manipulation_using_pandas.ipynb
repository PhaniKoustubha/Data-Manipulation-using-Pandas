**Sorting the data**

```python
# Sorting by a single column
sorted_df = df.sort_values(by='ChipRate')
print(sorted_df)
```

```python
# Sorting by multiple columns
sorted_df = df.sort_values(by=['BlowFlow', 'ChipRate'])
print(sorted_df)
```

```python
# Sorting in descending order
sorted_df = df.sort_values(by='ChipRate', ascending=False)
print(sorted_df)
```

```python
#sorting is ascending order
sorted_df = df.sort_values(by='ChipRate', ascending=True)
print(sorted_df)
```

```python
# Sorting rows based on index
sorted_df = df.sort_index(axis=0)
print(sorted_df)
```

```python
# Sorting columns based on column names
sorted_df = df.sort_index(axis=1)
print(sorted_df)
```

**Handling missing values**

```python
#Detecting Missing Values
missing_value = df.isnull()
print(missing_value)
```

```python
#Remove rows with missing values
df_drop = df.dropna()
print(df_drop)
```

```python
#Remove columns with missing values
df_drop = df.dropna(axis=1)
print(df_drop)
```

```python
#filling new value in place of null values
df_fill = df.fillna(value = 0)
print(df_fill)
```

```python
#handling missing value Conditionally
df_mean = df['ChipRate'].mean()
print(df_mean)
```

```python
df_fill = df.fillna(df_mean)
print(df_fill)
```
